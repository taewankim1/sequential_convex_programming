```python
import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib qt
%load_ext autoreload
%autoreload 2
import numpy as np
import time
import random
def print_np(x):
    print ("Type is %s" % (type(x)))
    print ("Shape is %s" % (x.shape,))
#     print ("Values are: \n%s" % (x))
```

```python
import sys
# sys.path.append('../')
sys.path.append('../')
sys.path.append('../model')
sys.path.append('../cost')
sys.path.append('../constraints')
sys.path.append('../utils')
import AircraftKinematicsModel
import AircraftKinematicsCost
import AircraftKinematicsConstraints
from scipy.integrate import solve_ivp
from Scvx import Scvx
from Scaling import TrajectoryScaling
```

```python
ix = 6
iu = 3
tf = 1000
N = 50
delT = tf/N
max_iter = 15
```

```python
myModel = AircraftKinematicsModel.AircraftKinematics('Hello',ix,iu,delT,linearization="numeric_central")
myCost = AircraftKinematicsCost.AircraftKinematics('Hello',ix,iu,N)
myConst = AircraftKinematicsConstraints.AircraftKinematics('Hello',ix,iu)

x_min = np.zeros(ix)
x_max = np.array([120*1e3,40*1e3,7000,240,2*np.pi,2*np.pi])
u_min = np.zeros(iu)
u_max = np.array([2*np.pi,2*np.pi,1126.3 * 1e3])

myScaling = TrajectoryScaling(x_min,x_max,u_min,u_max)
```

```python
# xi = np.array([-120*1e3,40*1e3,0,0])
xi = np.array([-120*1e3,40*1e3,7000,240,0,0])
xf = np.array([0,0,0,90,0,0])
x0 = np.zeros((N+1,ix))
for i in range(N+1) :
    x0[i] = (N-i)/N * xi + i/N * xf
u0 = np.zeros((N+1,iu))
u0[:,1] = 0
```

```python
i1 = Scvx('hi',N,max_iter,myModel,myCost,myConst,myScaling,
          type_discretization='foh',w_c=1,w_vc=1e4,w_tr=1e2,tol_vc=1e-4,tol_tr=1e-3,tol_bc=1,flag_policyopt=False)
x,u,xbar,ubar,total_num_iter,flag_boundary,l,l_vc,l_tr = i1.run(x0,u0,xi,xf)
```

```python
fig = plt.figure(1,figsize=(15,7))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax.plot(x[:, 0], x[:, 1], x[:, 2])
ax.plot(xbar[:, 0], xbar[:, 1], xbar[:, 2])
# ax.plot(xbar[:, 0], xbar[:, 1], xbar[:, 2],'o')
```

```python
plt.figure(figsize=(15,7))
plt.subplot(231)
plt.plot(x[:,0],x[:,1])
plt.plot(xbar[:,0],xbar[:,1])
plt.title('trajectory')
plt.xlabel('X')
plt.ylabel('Y')
plt.subplot(232)

plt.plot(x[:,4])
plt.title('gamma')
plt.subplot(233)
plt.plot(x[:,3])
plt.title('velocity')
plt.subplot(234)
plt.plot(ubar[:,0])
plt.title('gamma dot')
plt.subplot(235)
plt.plot(ubar[:,1])
plt.title('psi dot')
plt.subplot(236)
plt.plot(ubar[:,2])
plt.title('thrust')
```
