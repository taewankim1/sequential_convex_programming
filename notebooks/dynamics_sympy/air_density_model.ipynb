```python
rz = np.linspace(0,10000,10000)
```

```python
# flag_1 = rz < 11000 
T = 15.04 - 0.00649 * rz # celsius
p = 101.29 * np.power((T+273.1)/288.08,5.256)
rho1 = p / (0.2869 * (T + 273.1))
```

```python
rho2 = 1.226 * np.exp(-rz/9400)
```

```python
plt.figure(figsize=(15,5))
plt.plot(rz,rho1,color='tab:blue',label='original')
plt.plot(rz,rho2,'--',color='tab:red',label='exponential')
plt.legend()
```
